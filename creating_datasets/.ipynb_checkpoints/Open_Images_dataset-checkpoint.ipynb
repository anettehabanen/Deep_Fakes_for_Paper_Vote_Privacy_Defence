### Dataset analysis

```python
from PIL import Image
import os
import os.path as osp
import matplotlib.pyplot as plt
```

```python
open_images_path = "/home/ahabanen/fiftyone/open-images-v6/train/data" # Linux
#open_images_path = "C:/Users/ahabanen/fiftyone/open-images-v6/train/data" # Windows
file_names = os.listdir(open_images_path)
print(len(file_names))
```

```python
width_o = []
height_o = []

for i in range(len(file_names)):
    file_path = osp.join(open_images_path, file_names[i])
    image = Image.open(file_path)
    width, height = image.size
    width_o.append(width)
    height_o.append(height)
```

```python
plt.scatter(width_o, height_o)
plt.show()
```

```python
width_1 = []
height_1 = []

for i in range(len(file_names)):
    file_path = osp.join(open_images_path, file_names[i])
    image = Image.open(file_path)
    width, height = image.size
    if width < 1500 and height < 1500:
        width_1.append(width)
        height_1.append(height)
```

```python
plt.scatter(width_1, height_1)
plt.show()
```

```python
print("This dataset has " + str(len(file_names)) + " images.")
print("Images with the side lengths of (1024, 768): " + str(len(width_1)))
```

```python
width_2 = []
height_2 = []

for i in range(len(file_names)):
    file_path = osp.join(open_images_path, file_names[i])
    image = Image.open(file_path)
    width, height = image.size
    if (width != 1024 and height != 768) and (width != 768 and height != 1024):
        width_2.append(width)
        height_2.append(height)
```

```python
plt.scatter(width_2, height_2)
plt.show()
```

```python

count_w1024_h768 = 0

for i in range(len(file_names)):
    file_path = osp.join(open_images_path, file_names[i])
    image = Image.open(file_path)
    width, height = image.size
    if width == 1024 and height == 768:
        count_w1024_h768 += 1
```

```python
print("This dataset has " + str(len(file_names)) + " images.")
print("Images with the width 1024 and height 768: " + str(count_w1024_h768))
```
